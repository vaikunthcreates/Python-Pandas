```python
import pandas as pd
crime = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
crime.head(10)
```

```python
crime.info()
```

```python
crime['inyear']=pd.to_datetime(crime.Year, format = '%Y')
crime.head(4)
```

```python
del crime['Total']
crime.head(4)
```

```python
crime.set_index('inyear').head(3)
```

```python
crime['Decade'] = crime.Year //10
crime.head(15)
```

```python
agg = {c: 'sum' for c in crime.loc[:,'Population':'Vehicle_Theft'].columns}
agg['Population'] = 'last'
agg['Year'] = 'size'
crimed = crime.groupby('Decade').agg(agg)
print (crimed)
```

```python
# Most dangerous decade to live in, in US
trends = crimed.loc[:,'Violent':'Vehicle_Theft'].div(crimed.Year*crimed.Population,'rows')
print (trends)
```

```python
trends.plot()
```
